```python
# ✅ DYNAMIC LEARNING FEEDBACK SYSTEM (LLM-Driven)

from openai import OpenAI

# Step 1: Set up OpenRouter API client with DeepSeek Qwen3 8B model
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="sk-or-..."  # Replace with your real API key
)

### --------------------------------------
### STEP 2: Dynamic Task Generator
### --------------------------------------

def generate_learning_task(goal):
    prompt = f"""
A user is learning: \"{goal}\". 
Generate a short but meaningful writing or thinking task (e.g., opinion, argument, reflection) that can help assess their writing and thinking ability.
The task should allow us to evaluate vocabulary use, curiosity, confidence, compression, and emotional tone.
Only output the task as plain text.
"""

    response = client.chat.completions.create(
        model="deepseek/deepseek-r1-0528-qwen3-8b:free",
        messages=[
            {"role": "system", "content": "You are a helpful AI tutor."},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content

### --------------------------------------
### STEP 3: Score Response on Key Metrics
### --------------------------------------

def score_user_response(goal, task, user_response):
    prompt = f"""
The user is learning: \"{goal}\".
They were given this task: \"{task}\"
They responded with:
"""
{user_response}
"""

Score them from 0.0 to 1.0 on the following metrics:
1. Vocabulary Expansion (expressiveness)
2. Curiosity (depth of exploration)
3. Confidence (assertiveness and clarity)
4. Output Compression Ratio (clarity and conciseness)
5. Emotional Consistency (stable tone and intent)

Return a Python dictionary like this:
{
    "vocabulary_expansion_count": x,
    "curiosity_level": x,
    "confidence_level": x,
    "output_compression_ratio": x,
    "emotional_consistency_score": x
}
"""

    response = client.chat.completions.create(
        model="deepseek/deepseek-r1-0528-qwen3-8b:free",
        messages=[
            {"role": "system", "content": "You are an AI evaluator. Be concise and return only the Python dictionary."},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content

### --------------------------------------
### STEP 4: Generate Personalized Feedback
### --------------------------------------

def get_feedback_from_scores(goal, scores):
    prompt = f"""
The user is learning: \"{goal}\".
Here are their learning metric scores:
- Vocabulary Expansion Count: {scores['vocabulary_expansion_count']}
- Curiosity Level: {scores['curiosity_level']}
- Confidence Level: {scores['confidence_level']}
- Output Compression Ratio: {scores['output_compression_ratio']}
- Emotional Consistency Score: {scores['emotional_consistency_score']}

Please explain:
1. What this says about their current learning state
2. How they can improve
3. Practical next steps to focus on
"""

    response = client.chat.completions.create(
        model="deepseek/deepseek-r1-0528-qwen3-8b:free",
        messages=[
            {"role": "system", "content": "You are a helpful AI learning coach."},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content

### --------------------------------------
### STEP 5: Simulate Full Cycle
### --------------------------------------

# Example usage:
goal = "I want to learn how to write better persuasive essays."

# 1. Generate a task
task = generate_learning_task(goal)
print("\n--- GENERATED TASK ---\n", task)

# 2. Simulate user response (can be input from actual user)
user_response = "I think schools should allow students to pick their own schedules. When kids are forced into early mornings, they often struggle to focus. If we let teens pick their best times, they might do better."

# 3. Score response
scores_raw = score_user_response(goal, task, user_response)
print("\n--- METRIC SCORES ---\n", scores_raw)

# Convert string to dictionary
import ast
scores = ast.literal_eval(scores_raw)

# 4. Generate feedback
feedback = get_feedback_from_scores(goal, scores)
print("\n--- FEEDBACK ---\n", feedback)

# ✅ You can now loop this process or plug into Streamlit UI
```
